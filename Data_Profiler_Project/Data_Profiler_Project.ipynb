# 📊 PART D: EDA

```python
import matplotlib.pyplot as plt
df_csv['runs'].hist()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='runs', y='strike_rate', data=df_csv)
plt.title("Runs vs Strike Rate")
plt.show()
```

```python
corr = df_csv.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```
